```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
#import data
df =pd.read_csv("insurance.csv")
```

Use the data in the relevant columns to determine how age affects insurance costs

```python
# plot a scatter with age on the X-axis and charges on the Y-axis
x= df.iloc[:,:1].values
y= df.iloc[:,-1].values

plt.scatter(x,y)
plt.xlabel('age')
plt.ylabel('charges')
plt.show()

#alternative plotting
#plt.figure
#sns.scatterplot(data=df,
               # x='age',
                #y='charges')
```

```python
#using linear_model.LinearRegression() from sklearn, fit a model to your data, and make predictions on the data

age_model = LinearRegression()
age_model.fit(x,y)
```

```python
y_predict = age_model.predict(x)

#plot another scatter plot with the best fit line
plt.scatter(x,y)
plt.plot(x,y_predict, color ='r')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```
